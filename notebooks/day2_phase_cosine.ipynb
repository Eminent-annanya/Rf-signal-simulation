```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
A = 1
f = 5
fs = 100
t = np.linspace(0, 1, fs)

# Signals
sine_wave = A * np.sin(2 * np.pi * f * t)
cosine_wave = A * np.cos(2 * np.pi * f * t)

# Plot
plt.figure()
plt.plot(t, sine_wave, label="Sine Wave")
plt.plot(t, cosine_wave, label="Cosine Wave")

plt.title("Sine vs Cosine")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.legend()
plt.grid()

plt.show()
```

```python
# Phase shift
phi = np.pi / 4  # 45 degrees

shifted_wave = A * np.sin(2 * np.pi * f * t + phi)

# Plot
plt.figure()
plt.plot(t, sine_wave, label="Original Sine")
plt.plot(t, shifted_wave, label="Phase Shifted (45°)")

plt.title("Phase Shift Effect")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.legend()
plt.grid()

plt.show()
```

```python
phases = [0, np.pi/4, np.pi/2, np.pi]

plt.figure()

for phi in phases:
    wave = A * np.sin(2 * np.pi * f * t + phi)
    plt.plot(t, wave, label=f"Phase = {phi:.2f}")

plt.title("Different Phase Shifts")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.legend()
plt.grid()

plt.show()
```
